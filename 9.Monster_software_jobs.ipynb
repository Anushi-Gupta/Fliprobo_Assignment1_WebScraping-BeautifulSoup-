```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://www.monster.com/jobs/search/?q=software-developer&stpage=1&page=5")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
Jobs_Title=[]
Companies=[]
Job_location=[]
```

```python
#Jobs_Title of all the job listed in 5 pages.

jobs=soup.find_all('h2',class_='title')
for i in jobs:
    Jobs_Title.append(i.text.replace('\r\n',''))
    
Jobs_Title
```

```python
len(Jobs_Title)
```

```python
#Company_Names in 5 pages.

company=soup.find_all('div',class_='company')
for i in company:
    Companies.append(i.text.replace('\r\n','').replace('\n\n\n','').replace('\n',''))
    
Companies
```

```python
len(Companies)
```

```python
#Job_Location of all the job listed in webpage.

Location=soup.find_all('div',class_='location')
for i in Location:
    Job_location.append(i.text.replace('\r\n','').replace('\n',''))
    
Job_location
```

```python
len(Job_location)
```

```python
jobs=pd.DataFrame({})
```

```python
jobs['Job_Tiltle']=Jobs_Title[0:127]
jobs['Companies']=Companies[0:127]
jobs['Location']=Job_location[0:127]
```
